```python
import torch
import numpy as np
import torch.nn as nn
import random
train_size=80000
test_size=20000
m=15

def point(x,y):
    if (x<0)or(y<0)or(x>=m)or(y>=m):
        return 0
    return current_chess_board[x*m+y]

def add_chess(x,w=1):
    current_chess_board[x]=w
    
def check_chess_board():
    full=1
    tot=0
    ans=0
    for x in range(m):
        for y in range(m):
            if (point(x,y)==0):
                full=0
                for i in range(-4,1):
                    o1=o2=o3=o4=1
                    for j in range(i,i+5):
                        if j!=0:
                            if (point(x+j,y)!=1): o1=0
                            if (point(x+j,y+j)!=1): o2=0
                            if (point(x,y+j)!=1): o3=0
                            if (point(x+j,y-j)!=1): o4=0
                    if (o1==1)or(o2==1)or(o3==1)or(o4==1): 
                        ans=x*m+y
                        tot+=1
    if tot==0:
        return -1
    elif tot>1:
        return -2
    else:
        return ans

def print_chess_board():
    print(" \ ",end='')
    for x in range(m):
        print(chr(x+ord('a')),end=' ')
    for x in range(m):
        print()
        if x<9: print('0',end='')
        print(x+1,end=' ')
        for y in range(m):
            p=point(x,y)
            if p==0: print('.',end=' ')
            if p==1: print('o',end=' ')
            if p==-1: print('*',end=' ')
    print()
```

```python
# generate trainning and test set
train_set=np.empty((m*m,0))
test_set=np.empty((m*m,0))
train_label=np.zeros(train_size)
test_label=np.zeros(test_size)
i=0
while i<train_size+test_size:
    if i%100==0: print (i)
    current_chess_board=np.zeros((m*m,1))
    t=1
    add_chess((m//2)*m+m//2)
    for t in range(m*m):
#         print(t)
        current_chess_board=current_chess_board*-1
        x=check_chess_board()
        if x==-2 or t==m*m-1: break
        elif x!=-1:
            if (i<train_size):
#                 print_chess_board() 
#                 print(x)
                train_set=np.column_stack((train_set,current_chess_board))
                train_label[i]=x
            else:
                test_set=np.column_stack((test_set,current_chess_board))
                test_label[i-train_size]=x
            i+=1
            break
        for tt in range(10000):
            x=random.randint(0,m*m-1)
            if (current_chess_board[x]==0):
                current_chess_board[x]=1
                break
np.savetxt('trainning_set.txt',train_set,fmt='%d')
np.savetxt('test_set.txt',test_set,fmt='%d')
np.savetxt('trainning_label.txt',train_label,fmt='%d')
np.savetxt('test_label.txt',test_label,fmt='%d')
print ("done")
```

```python
#load data set
train_set=np.loadtxt('trainning_set.txt',dtype=int)
test_set=np.loadtxt('test_set.txt',dtype=int)
train_label=np.loadtxt('trainning_label.txt',dtype=int)
test_label=np.loadtxt('test_label.txt',dtype=int)
train_set=train_set.transpose()
train_set=train_set.reshape(train_size,1,m,m)
test_set=test_set.transpose()
test_set=test_set.reshape(test_size,1,m,m)
print ("done")
```

```python
import torch
from torch.autograd import Variable
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as opt
import random
batch_size=1024

class Net(nn.Module):
    def __init__(self):
        super(Net,self).__init__()
        self.conv1=nn.Conv2d(1,10,5)
#         self.conv2=nn.Conv2d(6,16,5)
        self.fc1=nn.Linear(10*5*5,64)
        self.dp=nn.Dropout(0.5)
        self.fc2=nn.Linear(64,64)
        self.fc3=nn.Linear(64,m*m)
        
    def forward(self,x):
        x=F.max_pool2d(F.relu(self.conv1(x)),2)
#         x = F.max_pool2d(F.relu(self.conv2(x)),2)
        x=x.view(-1,10*5*5)   #change
        x=self.fc1(x)
#         x=self.dp(x)
        x=F.relu(x)
        x = self.fc2(x)
#         x=self.dp(x)
        x=F.relu(x)
        x = self.fc3(x)
        return x        
    
def get_batch():
    ret0=np.empty((batch_size,1,m,m))
    ret1=np.empty((batch_size))
    x=0
    for i in range(batch_size):
        x=random.randint(0,train_size-1)
        ret0[i]=train_set[x:x+1]
        ret1[i]=train_label[x:x+1]
#     print(shape(ret0))
    global inputs,labels
    inputs=torch.from_numpy(ret0).float()
    labels=torch.from_numpy(ret1.astype('int64'))

def test(epoch):
    correct = 0
    total = 0
    for i in range(train_size):
        inputs=torch.from_numpy(np.array(train_set[i:i+1])).float()
        inputs=Variable(inputs)
#         print(inputs.size())
        outputs = net(inputs)
        _, predicted = torch.max(outputs.data, 1)
        total +=1
        correct += (predicted[0] == torch.from_numpy(train_label[i:i+1].astype('int64'))).sum()
#         print(predicted[0].int())
#         break
    print('Accuracy of epoch %d : the network on the 10000 train images: %d %%' % (epoch,100 * correct / total))
    return correct

lr=0.5
net=Net()
criterion=nn.CrossEntropyLoss()
optimizer=opt.SGD(net.parameters(),lr=lr,momentum=0.9)
last_loss=1e8 
# train_size=80000
last_test=0
for epoch in range(1000):
    running_loss=0
    for i in range(2*train_size//batch_size):
#         inputs=torch.from_numpy(np.array(train_set[i:i+batch_size])).float()
#         labels=torch.from_numpy(train_label[i:i+batch_size].astype('int64'))
#         i+=batch_size-1
        get_batch()
#         
#         print(inputs.size())
#         print(labels)
        inputs=Variable(inputs)
        labels=Variable(labels)
        optimizer.zero_grad()
        outputs=net(inputs)
#         break
        loss=criterion(outputs,labels)
        loss.backward()
        optimizer.step()
        running_loss+=loss.data[0]
    print("epoch=",epoch,end=' ')
    print("loss=",running_loss,end=' ')
    tmp=test(epoch)
    if (tmp<last_test and random.randint(0,2)==0): 
        lr=lr*0.5
        for param_group in list(optimizer.param_groups):
              param_group['lr'] = lr
        print(lr)
    last_test=tmp
    test2()
print ("done") 
```
